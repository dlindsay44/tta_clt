```python
"""
Write a program that
calculates the lengths of
sides of a triangle using
Pythagoras's Theorem.
Follow these instructions:
1. Print a menu for
Pythagoras's calculator
2. Prompt the user to
enter an option and
complete the calculation
"""
import math #Allows square root calculations

def pythagoras_fun_a(b, c): #Function to calculate and return leg a
    pythagoras_calc_a = math.sqrt(c**2 - b**2)
    return pythagoras_calc_a

def pythagoras_fun_b(a, c): #Function to calculate and return leg b
    pythagoras_calc_b = math.sqrt(c**2 - a**2)
    return pythagoras_calc_b

def pythagoras_fun_c(a, b): #Function to calculate and return hypotenuse
    pythagoras_calc_c = math.sqrt(a**2 + b**2)
    return pythagoras_calc_c

def pythagoras_proc(): #Procedure to take inputs, handle data type errors, put lengths into function, and print length
    option2 = "1" #Variable to run main menu (can return to main menu through final menu after calculation is finished)
    while option2 == "1": #'While' loop for main menu
        option = input("Welcome to the Pythagoras calculator. Please enter a number to choose a side's length to calculate: "
                    "1: Leg 'a'; 2: Leg 'b'; 3: Hypotenuse 'c'")
        
        if option == "1": #Takes inputs for b and c and solves for a. Prints all triangle lengths
            while True: #'While' loop to ensure hypotenuse longer than leg
                while True: #'While' loops to ensure valid data type
                    try:
                        leg_b = float(input("Enter leg 'b' length: "))
                        break
                    except ValueError:
                        print("Incorrect data type. Please enter leg 'b' length: ")
                while True:
                    try:
                        hyp_c = float(input("Enter hypotenuse 'c' length: "))
                        break
                    except ValueError:
                        print("Incorrect data type. Please enter hypotenuse 'c' length: ")
                if leg_b > hyp_c:
                    print("Error: hypotenuse must be longer than inputted leg")
                    continue
                else:
                    break
            length = pythagoras_fun_a(leg_b, hyp_c) #Puts variables into function
            print("Leg 'b' length = ", leg_b) #Prints lengths
            print("Hypotenuse 'c' length =", hyp_c) 
            print("Leg 'a' length =", length)
            while True: #'While' loop for final menu
                option2 = input("Calculation complete. Please choose an option: 1: Start new calculation; 2: Exit program") #Gives user the option to start new calculation or exit program
                if option2 == "1":
                    break
                elif option2 == "2":
                    print("Thank you for using the Pythagoras calculator. Goodbye")
                    break
                else:
                    print("Error: invalid option")
                    continue

        elif option == "2":  #Takes inputs for a and c and solves for b. Prints all triangle lengths
            while True: #'While' loop to ensure hypotenuse longer than leg
                while True: #'While' loop to ensure valid data type
                    try:
                        leg_a = float(input("Enter leg 'a' length: "))
                        break
                    except ValueError:
                        print("Incorrect data type. Please enter leg 'a' length: ")
                while True:
                    try:
                        hyp_c = float(input("Enter hypotenuse 'c' length: "))
                        break
                    except ValueError:
                        print("Incorrect data type. Please enter hypotenuse 'c' length: ")
                if leg_b > hyp_c:
                    print("Error: hypotenuse must be longer than inputted leg")
                    continue
                else:
                    break
            length = pythagoras_fun_b(leg_a, hyp_c) #Puts variables into function
            print("Leg 'a' length = ", leg_a)
            print("Hypotenuse 'c' length =", hyp_c)
            print("Leg 'b' length =", length) #Prints lengths
            while True: #'While' loop for final menu
                option2 = input("Calculation complete. Please choose an option: 1: Start new calculation; 2: Exit program") #Gives user the option to start new calculation or exit program
                if option2 == "1":
                    break
                elif option2 == "2":
                    print("Thank you for using the Pythagoras calculator. Goodbye")
                    break
                else:
                    print("Error: invalid option")
                    continue

        elif option == "3": #Takes inputs for a and b and solves for c. Prints all triangle lengths
            while True: #'While' loop to ensure valid data type
                try:
                    leg_a = float(input("Enter leg 'a' length: "))
                    break
                except ValueError:
                    print("Incorrect data type. Please enter leg 'a' length: ")
            while True:
                try:
                    leg_b = float(input("Enter leg 'b' length: "))
                    break
                except ValueError:
                    print("Incorrect data type. Please enter leg 'b' length: ")
            length = pythagoras_fun_c(leg_a, leg_b) #Puts variables into function
            print("Leg 'a' length = ", leg_a)
            print("Leg 'b' length =", leg_b)
            print("Hypotenuse 'c' length =", length) #Prints lengths
            while True: #'While' loop for final menu
                option2 = input("Calculation complete. Please choose an option: 1: Start new calculation; 2: Exit program") #Gives user the option to start new calculation or exit program
                if option2 == "1":
                    break
                elif option2 == "2":
                    print("Thank you for using the Pythagoras calculator. Goodbye")
                    break
                else:
                    print("Error: invalid option")
                    continue

        else:
            print("Error: invalid option")
            continue

pythagoras_proc() #Calls procedure

```
