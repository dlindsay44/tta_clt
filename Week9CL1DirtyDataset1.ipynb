```python
"""
1- Import the file named
Data Analyst_Sample.csv in Pandas
2- Explore the file with at least 3
common exploration methods in
Pandas
3- How many rows and columns the
data file has?
4- Which column names have missing
values and how many?
5- Remove missing values from
columns that contain Pandas object
datatype
6- Replace missing values by zero in
columns that contain numbers
7- Verify that no missing values are in
the data file
"""
#1- Import the file named Data Analyst_Sample.csv in Pandas

import pandas as pd #Allows Pandas to be used with 'pd' alias

das = pd.read_csv('Data_Analyst_Sample.csv') #Imports csv file as 'das'
```

```python
#2- Explore the file with at least 3 common exploration methods in Pandas

print('Max rows:', pd.options.display.max_rows) #Max rows system will return in full (60 by default), otherwise only headers and first and last 5 rows
print(das) #Prints dataframe
```

```python
print(das.tail(12)) #Prints last 12 rows of dataframe
```

```python
das.describe() #Shows basic descriptive statistics for numeric columns
```

```python
das.describe(include = object) #Descriptive statistice for objects
```

```python
#3- How many rows and columns the data file has?
#4- Which column names have missing values and how many?

print(das.info()) #Prints dataframe information 

#Number of rows and columns: 52481 entries (rows), 10 columns

#Column names with missing values (52481 - non-null count): 
#'0 Campaign Name': 36 rows with missing values
#'3 Date': 33 rows with missing values
#'4 Impressions': 30 rows with missing values
#'6 Cost': 23 rows with missing values
```

```python
"""
5- Remove missing values from
columns that contain Pandas object
datatype
"""

das2 = das.select_dtypes(include=['object']) #Only selects columns with datatype 'object' and puts them in new dataframe 'das2'
das2.dropna(inplace = True) #'dropna' removes empty rows, 'inplace = True' will not return new dataframe but remove all rows containing null values from original dataframe
print(das2.info())
```

```python
#6- Replace missing values by zero in columns that contain numbers
#7- Verify that no missing values are in the data file

das2 = das.select_dtypes(include=['number']) #Selects all number-based columns
das2.fillna(0, inplace = True) #'fillna' replaces empty values with '0'
print(das2.info()) #All non-null counts are the same as the number of rows, so no missing values in data file
```
